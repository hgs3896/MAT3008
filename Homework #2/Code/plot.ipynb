```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
import pandas as pd
```

```python
info = pd.DataFrame({
    "Bisection": [17,17,17],
    "Linear interpolation": [4,3,3],
    "Secant": [4,3,3],
    "Newton-Raphson": [3,3,3],
    "Bracketing Newton": [3,3,3],
    "Muller Method": [2,2,1],
})
```

```python
info = info.T
info.columns = ['2.404826', '5.520078', '8.653728']
```

```python
colors = ['r', 'g', 'b']
fig, ax = plt.subplots()
fig.patch.set_facecolor('white')
ax.set_title('Find the roots of the Bessel function')
fig.get_figure().set_size_inches(10, 5)
ax.set_xlabel('Method')
ax.set_ylabel('Number of iterations to finish')
ax.grid(axis='both')
for idx, col in enumerate(info):
    ax.scatter(info.index, info[col].values, c=colors[idx], alpha = 0.5, label=info.keys()[idx])
ax.legend()
```
